```python
LR_list = [5e-4, 1e-3]
HIDDEN_SIZES_list = [(32, 32), (64, 64)]

cfg_list = []
result_list = []

for lr,  hidde_sizes in product(LR_list, HIDDEN_SIZES_list):
    cfg = DQNConfig(
        LR=lr,
        HIDDEN_SIZES=hidde_sizes
    )
    cfg.variant = f"DQN_LR({lr})_HIDDEN_SIZES({hidde_sizes})"
    cfg_list.append(cfg)

print(cfg_list[-1].variant)
```

```python
for i, cfg in enumerate(cfg_list):
    print(f"Running experiment {i+1}: {cfg.variant}")
    result, _ = train_loop(cfg, variant=cfg.variant, output_mode="progress_bar")
    window = result.config.WINDOW
    final_avg = np.mean(result.returns[-window:])
    result.final_avg_return = final_avg
    print(f"Final avg return over last {window} episodes: {final_avg:.2f}")
    result_list.append(result)
```
