```python
import gymnasium as gym
from evaluation import evaluate_policy, show_metrics

def random_policy(env, obs):
    # `obs` is required by the interface but not used here.
    return env.action_space.sample()

metrics_random = evaluate_policy(random_policy, num_episodes=100, seed=0)

show_metrics("Random", metrics_random)
```

```python
import gymnasium as gym
from gymnasium.envs.box2d.lunar_lander import heuristic
from evaluation import evaluate_policy

def heuristic_policy(env, obs):
    # Use the official heuristic controller provided by Gymnasium.
    return heuristic(env, obs)

metrics_heuristic = evaluate_policy(heuristic_policy, num_episodes=100, seed=0)

show_metrics("Heuristic", metrics_heuristic)
```
